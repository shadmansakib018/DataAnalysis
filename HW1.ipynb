```python
# Q24. Write code to find the number of customers served every day. Sore the result in variable(s).
group_by_day_df = tips.groupby('day').count()
friday = group_by_day_df.iloc[0,0]
saturday = group_by_day_df.iloc[1,0]
sunday = group_by_day_df.iloc[2,0]
thursday = group_by_day_df.iloc[3,0]
total_customers_by_day = { "thursday":thursday, "friday":friday, "saturday":saturday, "sunday":sunday}
print(total_customers_by_day)
```

```python
total_customers_by_day.values()
```

```python
# Q25. Draw a line plot that shows number of customers served every day.
plt.plot(["Thursday",'Friday',"Saturday","Sunday"], total_customers_by_day.values())
plt.xticks([0.0,1.0,2.0,3.0],["Thursday",'Friday',"Saturday","Sunday"])
plt.xlabel('Day')
plt.ylabel('Number of Customers')
plt.title("Number of Customers Per Day")
plt.grid(True)
```

```python
# Q26. Use the groupby function to find the average tip per day.
# Store these averages in variables.
# These variables will be used next to display the averages in plots
avg_tip_per_day = tips.groupby('day').mean('numeric_only').tip
print(avg_tip_per_day.to_dict())
```

```python
# Q27. Draw a figure with 2 by 2 subplots. Each subplot shows a line plot of the tips given on a certain day.
# Show the average tip in each subplot.
# For example, these are shown as red asterisks in the diagram below. Give titles to the different subplots as well as a title to the whole figure.
# The line plt.figure(figsize=(8, 8)) can be used to specify figure dimensions in inches

tips_on_friday = tips[tips.day == "Fri"].tip
# tips_on_friday.reset_index(drop=True, inplace=True)

tips_on_thursday = tips[tips.day == "Thur"].tip
# tips_on_thursday.reset_index(drop=True, inplace=True)

tips_on_sunday = tips[tips.day == "Sun"].tip
# tips_on_sunday.reset_index(drop=True, inplace=True)

tips_on_saturday = tips[tips.day == "Sat"].tip
# tips_on_saturday.reset_index(drop=True, inplace=True)

fig = plt.figure(figsize=(8, 8))

axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot(tips_on_thursday)
axes1.plot(0,avg_tip_per_day['Thur'], marker='*', color='r')

axes2.plot(tips_on_friday)
axes2.plot(0,avg_tip_per_day['Fri'], marker='*', color='r')

axes3.plot(tips_on_saturday)
axes3.plot(0,avg_tip_per_day['Sat'], marker='*', color='r')

axes4.plot(tips_on_sunday)
axes4.plot(0,avg_tip_per_day['Sun'], marker='*', color='r')

axes1.set_title("Thursday")
axes2.set_title("Friday")
axes3.set_title("Saturday")
axes4.set_title("Sunday")

fig.suptitle("Tips Given on Different Days")
```

```python
# Q28 Draw a pie chart to compare the total tip amounts collected on the different days.
# Make sure the pie chart shows as a perfect square. Show percentages
total_tips = tips['tip'].sum()
percentage_tips_fri = tips_on_friday.sum()*100/total_tips
percentage_tips_thurs = tips_on_thursday.sum()*100/total_tips
percentage_tips_sat = tips_on_saturday.sum()*100/total_tips
percentage_tips_sun = tips_on_sunday.sum()*100/total_tips
data = [percentage_tips_thurs, percentage_tips_fri, percentage_tips_sat, percentage_tips_sun]
labels = ["Thur", "Fri", "Sat", "Sun"]
plt.pie(data, labels=labels, colors = ['red', 'cyan', 'orange', 'green'], autopct='%4.2f%%', )
plt.axis('equal')
```
